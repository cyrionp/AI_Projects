# Mean, Median ve Mod Örnekleri

```python
import numpy as np
```

```python
# Random kullanarak sample bir veri setini kendimiz oluşturalım
# Aylık gelir 5000 TL stardard deviation (standart sapma) 2000 TL olsun 1000 kişinin random aylık gelirini oluşturalım

gelirler = np.random.normal(5000,2000,1000)  # aylık_gelir, standard_deviation, kisi_sayisi
gelirler
```

```python
np.mean(gelirler)
```

```python
import matplotlib.pyplot as plt
plt.hist(gelirler, 100)
plt.show()
```

```python
np.median(gelirler)
```

```python
# Veri setine olağandışı bir veri girmek

gelirler = np.append(gelirler, [100000000])

np.mean(gelirler)
```

```python
np.median(gelirler)
```

```python
# Median verisi bu olağandışı veriye rağmen bozulmaz
# Median kontrol ederek olağandışı bir veri varsa fark edebiliriz
```

```python
# Mode örneği

yaslar = np.random.randint(7,18, size=300)

from scipy import stats
stats.mode(yaslar)
```

```python
np.median(yaslar)
```

```python
np.mean(yaslar)
```
