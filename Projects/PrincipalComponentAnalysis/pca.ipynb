# PCA ile Iris çiçeklerini tahmin etme

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

url = "pca_iris.data"
df = pd.read_csv(url, names=["sepal length","sepal width","petal length","petal width","target"])
df
```

```python
# Features x olarak ayrılsın ve target y olarak ayrılsın

features = ["sepal length","sepal width","petal length","petal width"]
x = df[features]
y = df[["target"]]
```

```python
# Standardizing

x = StandardScaler().fit_transform(x)
x
```

```python
# 4D verileri 2D'ye indirgemek

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data=principalComponents, columns=["principal component 1", "principal component 2"])
principalDf
```

```python
# Target columnu pca df sonuna eklemek

final_df = pd.concat([principalDf, df[["target"]]], axis=1)
final_df.head()
```

```python
# Çizdiriyorum

df_setosa = final_df[df.target=="Iris-setosa"]
df_virginica = final_df[df.target=="Iris-virginica"]
df_versicolor = final_df[df.target=="Iris-versicolor"]

plt.xlabel("pri comp 1")
plt.ylabel("pri comp 2")

plt.scatter(df_setosa["principal component 1"], df_setosa["principal component 2"], color="green")
plt.scatter(df_virginica["principal component 1"], df_virginica["principal component 2"], color="red")
plt.scatter(df_versicolor["principal component 1"], df_versicolor["principal component 2"], color="blue")
```

```python
# For loop ile daha iyi bir çizim

targets = ["Iris-setosa","Iris-versicolor","Iris-virginica"]
colors = ["g","b","r"]

plt.xlabel("pri comp 1")
plt.ylabel("pri comp 2")

for target, col in zip(targets,colors):
    df_temp = final_df[df.target==target]
    plt.scatter(df_temp["principal component 1"], df_temp["principal component 2"], color=col)
```
