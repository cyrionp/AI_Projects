```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# İlk seri x-ekseni, ikinci seri y-eksenidir
# Çizdirmek için plot kullanılır

plt.plot([1,2,3,4], [1,4,9,16])
```

```python
# Eksenleri variable olarak da tanımlayabiliriz

x = [1,2,3,4]
y = [1,4,9,16]

plt.plot(x,y)
plt.show()
```

```python
# Title ekleme

plt.title("My first graph!")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.show()
```

```python
# Otomatik olarak eksen çentiklerini yerleştirir ama biz de belirleyebiliriz

plt.xticks([1,2,3,4])  # İstersek küsüratları da gösterebilir
plt.yticks([1,4,9,16])
plt.title("My first graph!")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.show()
```

# Legend (Lejant) koyma

```python
plt.plot(x,y, label="x^2", color="green")
plt.xticks([1,2,3,4,5])
plt.yticks([1,4,9,16,25])
plt.title("My first graph!")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
# linewidth ekleme

plt.plot(x,y, label="x^2", color="blue", linewidth=2, linestyle="--", marker=".")
plt.xticks([1,2,3,4,5])
plt.yticks([1,4,9,16,25])
plt.title("My first graph!")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

# Aynı grafik üzerinde birden fazla çizim yapmak

```python
plt.plot(x,y, label="x^2", color="green", linewidth=2, linestyle="--", marker=".")
plt.xticks([1,2,3,4,5])
plt.yticks([1,4,9,16,25])
plt.title("My first graph!")
plt.xlabel("x axis")
plt.ylabel("y axis")

# 2. çizim
x2 = np.arange(0,5,0.5)
plt.plot(x2,x2*2, color="red", linewidth=2, marker=".", label="2*x")

plt.legend()
plt.show()
```

# Oluşturulan grafiği png olarak kaydetmek

```python
plt.plot(x,y, label="x^2", color="green", linewidth=2, linestyle="--", marker=".")
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,1,4,9,16,25])
plt.title("My first graph!")
plt.xlabel("x axis")
plt.ylabel("y axis")

# 2. çizim
x2 = np.arange(0,5,0.5)
plt.plot(x2,x2*2, color="red", linewidth=2, marker=".", label="2*x")

plt.legend()

plt.savefig("ilkgrafigim.png", dpi=300)  # dpi arttırarak grafik çözünürlüğü/büyüklüğünü arttırabiliriz

plt.show()
```

# Barchart

```python
x = ["Ankara", "İstanbul", "İzmir"]
y = [120,178,87]

plt.bar(x,y)
plt.show()
```

```python
# Farklı göstermek için

x = ["Ankara", "İstanbul", "İzmir"]
y = [120,178,87]

cubuklar = plt.bar(x,y)
cubuklar[1].set_hatch("/")
cubuklar[0].set_hatch(".")

plt.show()
```

# Detaylı örnekler

```python
gas = pd.read_csv("petrolfiyatlari.csv")

gas
```

```python
plt.title("Petrol Fiyatları")
plt.plot(gas["Year"], gas["USA"], "b-", label="USA")

plt.xlabel("Yıl")
plt.ylabel("Dolar")

plt.legend()
plt.show()
```

```python
plt.title("Petrol Fiyatları")
plt.plot(gas.Year, gas.USA, "b-", label="USA")
plt.plot(gas.Year, gas.Canada, "r-", label="Kanada")
plt.plot(gas.Year, gas["South Korea"], "g.-", label="Güney Kore")
plt.plot(gas.Year, gas.France, "y.-", label="Fransa")
plt.plot(gas.Year, gas.Germany, color="orange", label="Almanya")

plt.xlabel("Yıl")
plt.ylabel("Dolar")

plt.legend()
plt.show()
```

# Figure Size

Tablo büyüklüğü ayarlanabilir

```python
plt.figure(figsize=(14,5))
plt.title("Petrol Fiyatları")
plt.plot(gas.Year, gas.USA, "b-", label="USA")
plt.plot(gas.Year, gas.Canada, "r-", label="Kanada")
plt.plot(gas.Year, gas["South Korea"], "g.-", label="Güney Kore")
plt.plot(gas.Year, gas.France, "y.-", label="Fransa")
plt.plot(gas.Year, gas.Germany, color="orange", label="Almanya")

plt.xlabel("Yıl")
plt.ylabel("Dolar")

plt.legend()

plt.savefig("ikincigrafigim.png", dpi=300)

plt.show()
```
